```python
import pandas as pd
import json

# List of 40 signals with their PESTLE
signals = [
    ("Government Policy Announcements", "Political"),
    ("Cabinet/Parliament Decisions", "Political"),
    ("Government Sector Strike Warnings", "Political"),
    ("Police/Security Alerts", "Political"),
    ("Election-related Discussions", "Political"),
    ("Foreign Policy / International Agreements", "Political"),
    ("Tax Revision Rumors", "Political"),
    ("Public Protests & Demonstrations", "Political"),
    ("Inflation Mentions", "Economic"),
    ("Fuel Shortage Mentions", "Economic"),
    ("Dollar Rate Discussions", "Economic"),
    ("Tourism Search Trend (Google Trends)", "Economic"),
    ("Food Price Spikes", "Economic"),
    ("Stock Market Volatility", "Economic"),
    ("Foreign Investment News", "Economic"),
    ("Currency Black Market Mentions", "Economic"),
    ("Crime & Safety Alerts", "Social"),
    ("Public Sentiment (Social Media)", "Social"),
    ("Migration / Visa Interest", "Social"),
    ("Public Health Discussions", "Social"),
    ("Viral Social Trends", "Social"),
    ("Cultural Event Mentions", "Social"),
    ("Power Outages (CEB)", "Technological"),
    ("Telecom Outages", "Technological"),
    ("Cyberattack Mentions", "Technological"),
    ("E-commerce Growth Indicators", "Technological"),
    ("Digital Payments Failure Reports", "Technological"),
    ("New Regulations Affecting Businesses", "Legal"),
    ("Court Rulings Impacting Industries", "Legal"),
    ("Import/Export Restriction Changes", "Legal"),
    ("Customs/Port Delays", "Legal"),
    ("Rainfall Alerts", "Environmental"),
    ("Flood Warnings", "Environmental"),
    ("Heat Wave Alerts", "Environmental"),
    ("Landslide Warnings", "Environmental"),
    ("Cyclone Updates", "Environmental"),
    ("Air Quality Index Changes", "Environmental"),
    ("Drought Warnings", "Environmental"),
    ("Water Supply Cuts (NWSDB)", "Environmental"),
    ("Coastal Erosion / Tsunami Alerts", "Environmental")
]

# Map each signal to SWOT (basic reasoning: Threat, Weakness, Opportunity, Strength)
swot_map = {
    "Government Policy Announcements": "Opportunity",
    "Cabinet/Parliament Decisions": "Opportunity",
    "Government Sector Strike Warnings": "Threat",
    "Police/Security Alerts": "Threat",
    "Election-related Discussions": "Threat",
    "Foreign Policy / International Agreements": "Opportunity",
    "Tax Revision Rumors": "Threat",
    "Public Protests & Demonstrations": "Threat",
    "Inflation Mentions": "Threat",
    "Fuel Shortage Mentions": "Threat",
    "Dollar Rate Discussions": "Threat",
    "Tourism Search Trend (Google Trends)": "Opportunity",
    "Food Price Spikes": "Threat",
    "Stock Market Volatility": "Threat",
    "Foreign Investment News": "Opportunity",
    "Currency Black Market Mentions": "Threat",
    "Crime & Safety Alerts": "Threat",
    "Public Sentiment (Social Media)": "Opportunity",
    "Migration / Visa Interest": "Threat",
    "Public Health Discussions": "Threat",
    "Viral Social Trends": "Opportunity",
    "Cultural Event Mentions": "Opportunity",
    "Power Outages (CEB)": "Weakness",
    "Telecom Outages": "Weakness",
    "Cyberattack Mentions": "Threat",
    "E-commerce Growth Indicators": "Opportunity",
    "Digital Payments Failure Reports": "Weakness",
    "New Regulations Affecting Businesses": "Opportunity",
    "Court Rulings Impacting Industries": "Threat",
    "Import/Export Restriction Changes": "Threat",
    "Customs/Port Delays": "Weakness",
    "Rainfall Alerts": "Threat",
    "Flood Warnings": "Threat",
    "Heat Wave Alerts": "Threat",
    "Landslide Warnings": "Threat",
    "Cyclone Updates": "Threat",
    "Air Quality Index Changes": "Threat",
    "Drought Warnings": "Threat",
    "Water Supply Cuts (NWSDB)": "Weakness",
    "Coastal Erosion / Tsunami Alerts": "Threat"
}

# Create DataFrame
df = pd.DataFrame(signals, columns=["Signal", "PESTLE"])
df["SWOT"] = df["Signal"].map(lambda x: swot_map.get(x, "Other"))

# Save CSV
df.to_csv("signals_pestel_swot.csv", index=False)
print("✅ CSV saved as signals_pestle_swot.csv")

# Save JSON
signals_json = df.to_dict(orient="records")
with open("signals_pestel_swot.json", "w") as f:
    json.dump(signals_json, f, indent=4)
print("✅ JSON saved as signals_pestel_swot.json")
```
